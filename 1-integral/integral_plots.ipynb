```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
times_11 = np.loadtxt('integral_out.txt')
x = np.arange(1, 29)
times_11 = times_11[0] / times_11
```

```python
plt.figure(figsize=(10,6),dpi=300)
plt.scatter(x, times_11, c = 'b', label = '$N = 10^{11}$')
plt.plot(x, times_11, c = 'b', linestyle = '--', alpha = 0.3)
plt.grid(visible=True, which='major', color='black', linestyle='-', alpha = 0.3)
plt.grid(visible=True, which='minor', color='black', linestyle='--', alpha = 0.3)
# plt.legend()
plt.xlabel('N,     ')
plt.ylabel('$T_1/T_N$') 
plt.title('   ')
plt.minorticks_on()
plt.savefig('integral_plot_accel.png')
```
