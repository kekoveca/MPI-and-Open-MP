```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
times_11 = np.loadtxt('times_11.txt')
times_12 = np.loadtxt('times_12.txt')
x = np.arange(1, 29)
times_11 = times_11[0] / times_11
times_12 = times_12[0] / times_12
```

```python
plt.figure(figsize=(10,6),dpi=300)
plt.scatter(x, times_11, c = 'b', label = '$N = 10^{11}$')
plt.scatter(x, times_12, c = 'r', label = '$N = 10^{12}$')
plt.plot(x, times_11, c = 'b', linestyle = '--', alpha = 0.3)
plt.plot(x, times_12, c = 'r', linestyle = '--', alpha = 0.3)
plt.grid(visible=True, which='major', color='black', linestyle='-', alpha = 0.3)
plt.grid(visible=True, which='minor', color='black', linestyle='--', alpha = 0.3)
plt.legend()
plt.xlabel('N, количество элементарных отрезков равной длины')
plt.ylabel('$T_1/T_N$') 
plt.title('График ускорения параллельного алгоритма')
plt.minorticks_on()
plt.savefig('integral_plot_accel.png')
```
